# Een wiskundige verkenning van de insectenwereld

In deze lessenreeks verkennen we de insectenwereld op een wel zeer bijzondere manier: we gaan aan de slag met numerieke modellen om de grootte van populaties bladluizen, lieveheersbeestjes en kevers te beschrijven.

We zijn blij dat u de weg gevonden heeft naar onze verzameling notebooks, en lichten graag even toe aan wat u zich zoal kunt verwachten. Bij eventuele vragen kunt u zich steeds [via e-mail](mailto:jeroen.vanderhooft@ugent.be,natacha.gesquiere@ugent.be,francis.wyffels@ugent.be) wenden tot de auteurs!

## Doelgroep

De primaire doelgroep van deze lessenreeks is de 3de graad Wetenschappen-Wiskunde, al kan deze ook gebruikt worden voor de richtingen Economie-Wiskunde, Grieks-Wiskunde, Latijn-Wiskunde, en Technologische wetenschappen en Engineering. De reeks integreert wiskundige en programmatorische concepten, en komt op die manier tegemoet aan minimumdoelen uit de wiskunde en uit de informatiewetenschappen. Hieronder worden de minimumdoelen vermeld waar deze lessenreeks aan voldoet, samen met pedagogisch-didactische wenken die naar voren geschoven worden door het katholiek onderwijs enerzijds, en het gemeenschapsonderwijs anderzijds.

## Krachtlijnen katholiek onderwijs

### Wiskunde B+S'' - 3de graad D-finaliteit - D/2023/13.758

Hoewel alle vier de krachtlijnen in deze lessenreeks aan bod komen, steekt een er bovenuit: wiskundig modelleren en probleemoplossend denken. De leerlingen zullen namelijk:

- gebruik maken van wiskundige modellen in de vorm van grafieken, bijvoorbeeld om de populatiegroei te plotten doorheen de tijd
- gebruik maken van exponentiële en logistische groeimodellen om de populatiegroei te beschrijven doorheen de tijd
- gebruik maken van matrices om de populatiegroei te beschrijven doorheen de tijd
- gebruik maken van diagrammen zoals het fasediagram van het Lokta-Volterramodel
- fenomenen zoals de populatiegroei van insecten beschrijven aan de hand van wiskundige modellen

### Informaticawetenschappen - 3de graad D-finaliteit - D/2023/13.758

In deze lessenreeks wordt ingezet op het gebruik van numerieke methodes bij het oplossen van wiskundige problemen. Dit onderwerp komt vooral aan bod in hoofdstuk 4, waar de populatiegroei gemodeleerd wordt aan de hand van differentiaalvergelijkingen.
